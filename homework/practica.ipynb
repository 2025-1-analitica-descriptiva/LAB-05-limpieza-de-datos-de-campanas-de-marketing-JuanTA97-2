```python
import glob
import zipfile
import pandas as pd
import io
from pprint import pprint
```

```python
path = "C:/Github/Analitica_Descriptiva/Labs/LAB-05-limpieza-de-datos-de-campanas-de-marketing-JuanTA97-2/files/input"
files = glob.glob(f"{path}/*.zip")
for file in files:
    with zipfile.ZipFile(file, 'r') as fzip:        
        for name in fzip.namelist():
            with fzip.open(name) as fcsv:
                df = pd.read_csv(io.TextIOWrapper(fcsv, encoding='utf-8'))
                print(df.head())
```
